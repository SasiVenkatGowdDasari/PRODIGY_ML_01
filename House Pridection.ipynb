# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

# Load the Data

```python
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')
```

# Select Relevant Features

```python
features=['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']
```

```python
X_train = train_df[features]
y_train = train_df['SalePrice']
```

# Handling Missing Values

```python
X_train.isnull().sum()
```

```python
test_df.isnull().sum()
```

```python
test_df.fillna(0,inplace=True)
```

# Add a feature for total bathrooms in both training and testing sets

```python
# X_train['TotalBath'] = X_train['FullBath'] + 0.5 * X_train['HalfBath']
# X_train = X_train.drop(columns=['FullBath', 'HalfBath'])

X_train.loc[:, 'TotalBath'] = X_train['FullBath'] + 0.5 * X_train['HalfBath']
X_train.drop(columns=['FullBath', 'HalfBath'], inplace=True)
```

```python
X_train.head()
```

```python
X_test = test_df[features].copy()  # Use copy to avoid the SettingWithCopyWarning
X_test['TotalBath'] = X_test['FullBath'] + 0.5 * X_test['HalfBath']
X_test.drop(columns=['FullBath', 'HalfBath'], inplace=True)

print(X_train.head())
print(X_test.head())
```

# Standardize the features

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Train the Linear Regression Model

```python
model=LinearRegression()
```

```python
model.fit(X_train_scaled,y_train)
```

# Predict on the training set to evaluate the model

```python
y_train_pred = model.predict(X_train_scaled)
```

# Calculate evaluation metrics

```python
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_train_pred)

print(f'Training Set Mean Absolute Error: {mae_train}')
print(f'Training Set Mean Squared Error: {mse_train}')
print(f'Training Set Root Mean Squared Error: {rmse_train}')
print(f'Training Set RÂ² Score: {r2_train}')
```

```python
# Make predictions on the test set
test_predictions = model.predict(X_test_scaled)

```

```python
# Create a DataFrame with the predictions
test_results = pd.DataFrame({
    'Id': test_df['Id'],
    'SalePrice': test_predictions
})
```

```python

# Save the results to a CSV file
test_results.to_csv('test_predictions.csv', index=False)
print("Predictions saved to test_predictions.csv")
```
